```python
#如何用NumPy实现蒙特卡洛模拟股票价格

import numpy as np

def stock_monte_carlo(S0=100, mu=0.08, sigma=0.2, n_sim=1000, n_days=252):
    dt = 1 / 252
    drift = (mu - 0.5 * sigma**2) * dt
    diffusion = sigma * np.sqrt(dt)
   
    Z = np.random.standard_normal((n_sim, n_days))
    log_ret = drift + diffusion * Z
    cum_log = np.cumsum(log_ret, axis=1)
    price_path = S0 * np.exp(cum_log)
 
    full_path = np.hstack([np.full((n_sim, 1), S0), price_path])
    return full_path

paths = stock_monte_carlo()
print("模拟路径维度(路径数,总天数+初始价):", paths.shape)
```
